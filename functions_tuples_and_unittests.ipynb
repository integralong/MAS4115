pytest is a minimalistic (but poweful if needed) frameworkk of testing python code, see blow for an example

```python
# we can place asserts in our code to make sure something is true
assert abs(-1) == 1
```

### Tuples
in python tuples are used to store a (small) bunch of values of possbile different type. very useful for returning more than one element from a function.


```python
def give_three_numbers():
    return 1, 2, 3.14 # this creates and returns a tuple
```

```python
t = give_three_numbers() # t is tuple
print(type(t)) # type of t = tuple
print(t)
print(t[0] + t[1] + t[2]) # we can index elements of the tuple

a, b, c = give_three_numbers() # automatic unpacking of a tuple!
print(a, b, c) 

first, *rest = give_three_numbers() # slightly more advanced
```

here, you might wonder why print(t[0] + t[1] + t[2]) does not print 6.14 but print 6.140000000000001.
computer store floating-point numbers in binary, not decimal.

* decimal fractions -> base-10
* binary fractions -> base-2
some decimals are impossible to represent exactly in base-2

so python stores the closest representable binary number, not the exact decimal value.

also, for the difference between 
**a, b, c = give_three_numbers()** and **first, *rest = give_three_numbers()** is 
first one should care about the count, otherwise it gets value error
however, the second one is care only one variable may use * to capture.
all remaining elements into a list. so, counts don't have to match exactly.

### Task 0

Write a function which takes to vectors a, b in R^2 and returns a vevtor t such that if we add t to a, we will ge b. 

We can represent each vector as a tuple (of size 2)

Write a coiple of unit tests and run them!

first, i tried with abs(b-a) however, then a and b cannot be vector(tuple). a b are scalars.
so b - a is unavailable in tuple.

correct way is using tuple! + index

```python
%%file x.py

def sub(a,b):
    '''
    Remember that both a and b are vectors, represented as tuples (a-b)
    '''
    return (b[0] - a[0], b[1] - a[1])


def test_sub_zero():
    zero_vector = (0,0)
    assert sub(zero_vector, zero_vector) == zero_vector

def test_sub_positive():
    a = (4, 2)
    b = (6, 4)
    assert sub(a,b) == (2,2)

def test_sub_negative():
    a = (-1, 1)
    b = (-2, 3)
    assert sub(a,b) == (-1,2)


```

### Default arguments
it's often convenient to have default values to some of the arguments
if we want to override an default argument value, it's often clearer if we explicitly spell out the name.

```python
# 1e-8 means 10**-8, see https://en.wikipedia.org/wiki/Scientific_notation
def are_epsilon_close(a, b, eps=1e-8):
  return abs(a-b) <= eps

assert are_epsilon_close((2**0.5)**2, 2) # using the default eps
assert are_epsilon_close((2**0.5)**2, 2, eps=1e-12) # overriding eps
# assert are_epsilon_close((2**0.5)**2, 2, eps = 0) # fails!
```
