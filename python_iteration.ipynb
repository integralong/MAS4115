### Task 2
write a function divisors which returns all the divisors which return all the divisors of an integer n

```python
def divisors(n):
    divs = []
    for x in range (1, n+1):
        if n % x == 0:
            elements = divs.append(x)
    return divs

elements = divisors (24)
print(elements)
    
```

```python
# fancy way
def divisors(n):
    return [x for x in range(1, n+1) if n%x == 0]

divis= divisors(24)
print(divis)
```

### Sorting (super useful, but error prone)

after sorting a list, its elements are in non-decreasing order

two functions:
* list.sort (sorts a list, returns nothing)
* sorted (returns a sorted copy, does not modify the list)

```python
L = [-3, 2, -4, 3]
print("L before:\t", sorted(L))
print("sorted returns:\t", sorted(L))

print("L after sorted:\t", L)
L.sort()
print("L after sort:\t", L)
```

```python
# given an integer  ğ‘›  decide if it's a prime number

L = [1,2,3,4,5,6]
def find_prime_number(L):
    for x in L:
        if x < 2:
            print(x, "is not a prime number")
            continue
        is_prime = True
        for i in range(2, x):
            if x % i == 0:
                is_prime = False
                break
        if is_prime:
            print(x, "is a prime number")
        else:
            print(x, "is not a prime nubmer")
            
find_prime_number(L)
        
```

```python
# given a number  ğ‘›  compute how many numbers in between 0 and  ğ‘›  (including  ğ‘› )
# are perfect squares (namely the square of some natural number; 
# e.g. 1 and 25 are perfect squares; 2 and 26 sadly aren't).

import math


def find_perfect_squares(numbers):
    perfect_squaes = []
    for x in numbers:
        root = int(math.sqrt(x))
        if root * root == x:
            perfect_squaes.append(x)
    return perfect_squaes

n = 10
L = list(range(1, n+1))
L1 = find_perfect_squares(L)
print(L)
print(len(L))

print(L1)
print(len(L1))
```
