### Overwrite

In python, you overwrite a variable simply by using the assignment operator (=) again with a new value. This changes the reference of the variable name to a new object in memory.

### basic oberwriting

```python
# initial assignment
x = 10
print(x) # output: 10

print(type(x))

# overwriting the varaible with a new value and different type
x = 'hello'
print(x) # output: hello

print(type(x))
```

### updating variables

```python
# initial assignmjent
number = 5
print(number) # output: 5

# updating using compound assignment (+=)
# this is equivalent to: number = number + 1
number += 1
print(number) # output: 6

# updating using comound assignment (*=)
# this is equivalent to: number = number * 2
number *= 2
print(number)
```

other compound operators include -=, /=

### overwriting within functions

by default, assigning a new value to variable inside a function creates a new local variables. To modify a global variable from within a function, you must use the global keyworkd.

```python
# Global variable
value = 10

def overwrite_loal():
    # this creates a new local varaible named 'value'
    value = 20
    print(f"Inside function (local): {value}")


def overwrite_global():
    # this modifies the global varialbe 'value'
    global value
    value = 30
    print(f"Inside function (global): {value}")

overwrite_loal()
print(f"Outside function after local overwrite: {value}")

overwrite_global()
print(f"Outside function after global overwrite: {value}")
```
