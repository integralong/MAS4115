### Updating variables

```python
t = 0 # starts with integer
print(type(t))
print(t)
x = 5

t = t + 10*x # update syntax
print(t)

t += 5 # t = t + 5
print(t)

t *= 2 # t = t * 2
print(t)

t /= 2 # t = t / 2
print(t)
```

```python
print(type(t))
```

### Multi-initialization and swapping
(very useful)

```python
a, b = 5, 10 # simply initializes a as 5 and b as 10
print(a, b)
a, b = b, a # idiomatic swap
print(a, b)

c = b, a 
print(c)

d = a, b
print(d)
```
