About "n -> infinity" vs "Δx→0"
They are the same idea, just phrased differently.

You split [a,b] into n equal pieces, so
dx = (b-a)/n

* if n -> infinity, then authomatically dx -> 0
* if dx -> o, that means you're using finer and finer partition, which implies n -> infinity.
so both implication directions hold.

In code here, we have num_elements, which pick a finite n. Increasing n makes dx smaller, which approximates that limit.

Therefore, although the integral is defined as a limit as Δx→0, numerically this limit must be realized by increasing the number of subintervals n, since shirinking Δx alone would require infinitely many funcion evaluations.
